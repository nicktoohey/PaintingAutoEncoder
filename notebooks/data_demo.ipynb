```python
import sys
sys.path.append('../')

import torch
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
from dataset.dataloader import get_data_loaders
import numpy as np
import matplotlib.pyplot as plt
```

```python
example_loader = get_data_loaders(name='example', batch_size=32)
#train_loader, test_loader, val_loader = get_data_loaders(base_path='../data', name='all', batch_size=32)
```

```python
import matplotlib.pyplot as plt


images, labels = next(iter(example_loader))
#images, labels = next(iter(train_loader))


def denormalize(tensor):
    return tensor * 0.5 + 0.5


fig, axes = plt.subplots(1, 5, figsize=(15, 3))
for i, ax in enumerate(axes):
    img = denormalize(images[i]).permute(1, 2, 0).numpy()
    ax.imshow(img)
    ax.set_title(labels[i])
    ax.axis('off')
plt.show()
fig.savefig('../assets/example_batch.png')
```
